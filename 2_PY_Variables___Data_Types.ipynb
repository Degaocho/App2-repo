## Variables: The building blocks of Python

### What is a variable?
We saw in the previous module that Python is one heck of a calculator. But you can imagine how more complex calculations would require us to remember a lot of values. That's where ***variables*** come in!  

A variable is a storage unit for different values like numbers or text. Variables must be specific and case sensitive because you call up the value of a variable by using its name. 



# Guidelines for naming variables

- [MUST] Variables names cannot start with special characters or symbols (e.g. @, #, $, numbers, %, -, =, *, etc), underscore is allowed especially between words.


- [MUST] You cannot have special characters or symbols in the name of a variable except numbers can appear at the end of of a variable name


- [Optionally] Ensure the name of a variable describes the content it holds or the operation it is used to carry out.

````python
# Remember this cell from the previous module? Meet ints. 

# Guidelines for naming variables


x = 40.75

y = 17

x + y # x and y are operands, and + is the operator
````

````python
type(x)
````

````python
type(y)
````

In the above cell we created (or *declared*) two variables: *x* and *y*. We did this by simply using an **=** sign much like we would in algebra.  
In this example we only used numerical values (these are called ***integers*** or ***ints***). So what happens when you use text values?


````python
# These are called strings:

z = 'Olouge is 50 years old. '
w = "Olouge is running late, again!"

z + w
````

````python
print(type(z))
print(type(w))
````

Something different happened!

When **ints** are added together, Python returns their sum as another int. Text values are referred to as ***strings***. These can be denoted using 'single' or "double" quotation marks as you can see above.  

When strings are added together, that's called *concatenation*. They're basically just pushed up against together. But remember, spaces are also counted. If you notice, there's no space at the end of the first string. Try adding a space after "is" and see what happens.  

This is a key principle of Python. *Different* types have *different* behavior.   




Let's try something a bit more realistic. 

Let's say we're working with personal financial data and we're trying to figure out a budget.

````python
# Let's put some numbers together:

# Income: $
monthly_income = 5000

# Taxes:
tax_rate = 0.184

# Expenses
rent = 1200
food = 320
wifi = 100
transport = 300
misc = 400
````

````python
type(tax_rate)
````

````python
type(monthly_income)
````

Notice how that cell didn't have an output? Even though we ran the cell, it had no commands other than assigning values to different variables. If we tried to add something (like in the original example), then there would be an output.

Ok. We've created seven variables and assigned each a different numeric value to represent our income, taxes, and expenses for a given month. Let's see how much we have left over after we pay Uncle Sam and cover our expenses. 

````python
# Income, after taxes:
AT_income = monthly_income * (1 - tax_rate)

print(AT_income)
````

````python
# Total expenses
expenses = rent + food + wifi + transport + misc

print(expenses)
````

````python
# What's left?

AT_income - expenses
````

Not bad. Assuming we saved all of that every month, how long would it take us to have six months of expenses covered in case of an emergency?

````python
# Let's declare a new variable to store that value:
max_savings = AT_income - expenses

# Now let's calculate how long it would take: 

# 2320 of expense per month  * 6 months = 13,920 this is emergency funds

emergency_funds_worth_six_months_expense = expenses * 6

months_to_save_for_planned_emergency = emergency_funds_worth_six_months_expense / max_savings

print("We need {f} months".format(f = months_to_save_for_planned_emergency))

print(str(months_to_save_for_planned_emergency)+" months")
````

Huh...nearly eight months! Guess we better start saving then.

## Mini Project 1:
You are working for Etech consulting as a python programmer. A new project for a banking system just came in. Your team manager asked you to write a python script that will calculate the total amount due for a loan after borrowing amount amt for n months with an interest rate of r%

The formula for compound interest rate calculation is: 
image.png

Where P is the starting principle amount. For example, if you borrowed $10,000 for 36 months at an interest rate of 5%, the amount after 36 months will be:

image.png



````python
# P*1 + r ^n/12
````

````python
2**3 # ** raises a number on the left to the power of the number on the right
````

````python
# Solution: Mini Project 1

# Variables: P, r, n
P = 10000
r = 5/100
n = 36

#Auxiliary variables: power, rate_sum, raised
power = n/12
rate_sum = 1 + r
raised = rate_sum ** power

# Intermediate variables: Total_amount
# Total_amount = P * ( (1+r) ** (n/12) )

Total_amount = P * raised


print(Total_amount)
````

## Mini Project 2:
A construction firm hired you to implement a python code that allows them accurately know the value of the area of circular shapes when constructing houses, write the python script to allow them get that done.

The area of a circle is pi times the radius squared (A = π r²)

Where π = 3.14 to 2 decimal places

````python
# Solution: Mini Project 2

# Variables: PI, r
PI = 3.14
r = 5 # Assuming the sample circular shape is having a radius of 2
POWER = 2

# Auxiliary variable: raised, power
raised = r ** POWER # π * r²


# Intermediate variables: Area
Area = PI * raised

print(Area)
````
